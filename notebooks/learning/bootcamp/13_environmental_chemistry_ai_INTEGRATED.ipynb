## Section 2: Advanced Environmental Monitoring

Dive deeper into environmental monitoring across different media:

```python
# Test different monitoring systems
monitoring_types = ["air_quality", "water_quality", "soil_contamination"]

monitoring_results = {}

for monitor_type in monitoring_types:
    print(f"\nüîç Analyzing {monitor_type.replace('_', ' ').title()}...")
    
    # Initialize monitoring system
    monitor = EnvironmentalMonitoringSystem(monitor_type)
    
    # Generate and analyze data
    data = monitor.generate_sample_data(1200)
    metrics = monitor.train_pollution_predictor(data)
    importance = monitor.get_feature_importance()
    
    monitoring_results[monitor_type] = {
        "metrics": metrics,
        "top_factors": importance.head(3).to_dict('records')
    }
    
    print(f"   ‚úÖ Model R¬≤ Score: {metrics['r2']:.3f}")
    print(f"   üìä Parameters monitored: {len(monitor.parameters)}")
    print(f"   üéØ Top factor: {importance.iloc[0]['parameter']} ({importance.iloc[0]['importance']:.3f})")

# Compare monitoring systems
print("\nüìä MONITORING SYSTEM COMPARISON:")
print("=" * 50)
for monitor_type, results in monitoring_results.items():
    r2_score = results["metrics"]["r2"]
    top_factor = results["top_factors"][0]["parameter"]
    emoji = "ü•á" if r2_score == max([r["metrics"]["r2"] for r in monitoring_results.values()]) else "üìà"
    print(f"{emoji} {monitor_type.replace('_', ' ').title()}: R¬≤={r2_score:.3f}, Key Factor: {top_factor}")

# Demonstrate real-time prediction
print("\nüîÆ Real-time Pollution Prediction Demo:")
air_monitor = EnvironmentalMonitoringSystem("air_quality")
air_data = air_monitor.generate_sample_data(1000)
air_monitor.train_pollution_predictor(air_data)

# Simulate new measurements
new_measurements = air_data.sample(5)[air_monitor.parameters]
predictions = air_monitor.predict_pollution(new_measurements)

for i, (idx, row) in enumerate(new_measurements.iterrows()):
    print(f"   üìç Location {i+1}: Predicted pollution index = {predictions[i]:.2f}")
    status = "üü¢ Good" if predictions[i] < 50 else "üü° Moderate" if predictions[i] < 100 else "üî¥ Unhealthy"
    print(f"     Status: {status}")
```

## Section 3: Green Chemistry Optimization

Optimize chemical processes for sustainability using AI:

```python
# Initialize Green Chemistry Optimizer
green_optimizer = GreenChemistryOptimizer()

# Generate and analyze reaction data
reaction_data = green_optimizer.generate_reaction_data(800)
print(f"üß™ Generated data for {len(reaction_data)} chemical reactions")
print(f"‚ôªÔ∏è Green metrics evaluated: {', '.join(green_optimizer.green_metrics)}")

# Optimize reaction conditions
optimization_results = green_optimizer.optimize_reaction_conditions(reaction_data)

print("\nüéØ GREEN CHEMISTRY OPTIMIZATION RESULTS:")
print(f"\n‚ö° Optimal Reaction Conditions:")
optimal = optimization_results["optimal_conditions"]
print(f"   ‚Ä¢ Temperature: {optimal['temperature']:.1f}¬∞C")
print(f"   ‚Ä¢ Pressure: {optimal['pressure']:.1f} atm")
print(f"   ‚Ä¢ Catalyst Loading: {optimal['catalyst_loading']:.2f}%")
print(f"   ‚Ä¢ Solvent Polarity: {optimal['solvent_polarity']:.1f}")
print(f"   ‚Ä¢ Reaction Time: {optimal['reaction_time']:.1f} hours")
print(f"   ‚Ä¢ Substrate Concentration: {optimal['substrate_concentration']:.2f} M")

print(f"\nüåü Maximum Green Score Achieved: {optimization_results['max_green_score']:.1f}")
print(f"üìä Model Performance (R¬≤): {optimization_results['model_score']:.3f}")

print(f"\nüìà Most Important Parameters for Green Chemistry:")
importance_df = optimization_results["feature_importance"]
for i, row in importance_df.head(6).iterrows():
    print(f"   {i+1}. {row['parameter'].replace('_', ' ').title()}: {row['importance']:.3f}")

# Test different reaction conditions
print("\nüß™ Testing Alternative Reaction Conditions:")
test_conditions = [
    {"temperature": 50, "pressure": 1.5, "catalyst_loading": 2.0, 
     "solvent_polarity": 3.0, "reaction_time": 4.0, "substrate_concentration": 1.0},
    {"temperature": 150, "pressure": 5.0, "catalyst_loading": 0.5, 
     "solvent_polarity": 8.0, "reaction_time": 12.0, "substrate_concentration": 0.5},
    {"temperature": 100, "pressure": 3.0, "catalyst_loading": 1.5, 
     "solvent_polarity": 5.5, "reaction_time": 8.0, "substrate_concentration": 0.8}
]

for i, conditions in enumerate(test_conditions, 1):
    prediction = green_optimizer.predict_green_metrics(conditions)
    score = prediction["predicted_green_score"]
    print(f"   üî¨ Condition Set {i}: Green Score = {score:.1f}")
    
    # Categorize performance
    if score > 70:
        category = "üåü Excellent"
    elif score > 60:
        category = "‚úÖ Good"
    elif score > 50:
        category = "‚ö†Ô∏è Moderate"
    else:
        category = "‚ùå Poor"
    
    print(f"     Performance: {category}")

# Green chemistry insights
print("\nüí° GREEN CHEMISTRY INSIGHTS:")
print(f"   ‚Ä¢ Optimal temperature range appears to be around {optimal['temperature']:.0f}¬∞C")
print(f"   ‚Ä¢ Lower catalyst loading ({optimal['catalyst_loading']:.1f}%) can be more sustainable")
print(f"   ‚Ä¢ Reaction time optimization can significantly impact green score")
print(f"   ‚Ä¢ Pressure and solvent choice are critical for sustainability")
```

## Section 4: Atmospheric Chemistry Analysis

Analyze atmospheric chemistry data and forecast air quality:

```python
# Initialize Atmospheric Chemistry Analyzer
atm_analyzer = AtmosphericChemistryAnalyzer()

# Generate atmospheric time series data
atm_data = atm_analyzer.generate_atmospheric_data(2000)  # ~3 months of hourly data
print(f"üå¨Ô∏è Generated atmospheric data: {len(atm_data)} hourly measurements")
print(f"üß™ Trace gases monitored: {', '.join(atm_analyzer.trace_gases)}")
print(f"üìÖ Time range: {atm_data['timestamp'].min()} to {atm_data['timestamp'].max()}")

# Analyze atmospheric trends
trend_analysis = atm_analyzer.analyze_atmospheric_trends(atm_data)

print("\nüìä ATMOSPHERIC CHEMISTRY TREND ANALYSIS:")
print("=" * 55)

for gas, trends in trend_analysis.items():
    print(f"\nüß™ {gas}:")
    print(f"   ‚Ä¢ Mean Concentration: {trends['mean_concentration']:.2f} ppm")
    print(f"   ‚Ä¢ Standard Deviation: {trends['std_concentration']:.2f} ppm")
    print(f"   ‚Ä¢ Hourly Trend: {trends['trend_per_hour']:.6f} ppm/hour")
    print(f"   ‚Ä¢ Daily Peak: {trends['daily_peak_hour']}:00")
    print(f"   ‚Ä¢ Seasonal Peak: Month {trends['seasonal_peak_month']}")
    
    # Trend interpretation
    if abs(trends['trend_per_hour']) > 0.001:
        trend_direction = "üìà Increasing" if trends['trend_per_hour'] > 0 else "üìâ Decreasing"
        print(f"   ‚Ä¢ Long-term Trend: {trend_direction}")
    else:
        print(f"   ‚Ä¢ Long-term Trend: ‚û°Ô∏è Stable")

# Air quality forecasting
print("\nüîÆ AIR QUALITY FORECASTING:")
forecasts = atm_analyzer.forecast_air_quality(atm_data, hours_ahead=24)

for gas, forecast in forecasts.items():
    value = forecast["forecast_value"]
    accuracy = forecast["accuracy_r2"]
    conf_low, conf_high = forecast["confidence_interval"]
    
    print(f"\nüìä {gas} (24h forecast):")
    print(f"   ‚Ä¢ Predicted Value: {value:.2f} ppm")
    print(f"   ‚Ä¢ Confidence Interval: {conf_low:.2f} - {conf_high:.2f} ppm")
    print(f"   ‚Ä¢ Model Accuracy (R¬≤): {accuracy:.3f}")
    
    # Air quality assessment
    if gas in ["NO2", "SO2", "CO"]:
        if value < 20:
            quality = "üü¢ Good"
        elif value < 50:
            quality = "üü° Moderate"
        else:
            quality = "üî¥ Unhealthy"
    else:
        quality = "üìä Within normal range"
    
    print(f"   ‚Ä¢ Air Quality: {quality}")

# Seasonal analysis
print("\nüóìÔ∏è SEASONAL PATTERNS SUMMARY:")
seasonal_summary = {}
for gas, trends in trend_analysis.items():
    seasonal_summary[trends['seasonal_peak_month']] = seasonal_summary.get(trends['seasonal_peak_month'], []) + [gas]

month_names = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
for month, gases in sorted(seasonal_summary.items()):
    print(f"   ‚Ä¢ {month_names[month-1]}: Peak concentrations for {', '.join(gases)}")

# Weather correlation analysis
print("\nüå§Ô∏è WEATHER-POLLUTION CORRELATION:")
import pandas as pd
correlation_matrix = atm_data[['temperature', 'wind_speed', 'humidity'] + atm_analyzer.trace_gases].corr()

weather_vars = ['temperature', 'wind_speed', 'humidity']
for weather in weather_vars:
    print(f"\nüå°Ô∏è {weather.title()} correlations:")
    correlations = correlation_matrix[weather][atm_analyzer.trace_gases].sort_values(key=abs, ascending=False)
    for gas, corr in correlations.head(3).items():
        direction = "üìà Positive" if corr > 0 else "üìâ Negative"
        strength = "Strong" if abs(corr) > 0.5 else "Moderate" if abs(corr) > 0.3 else "Weak"
        print(f"   ‚Ä¢ {gas}: {corr:.3f} ({direction}, {strength})")
```

## Section 5: Integrated Environmental Workflow

Demonstrate the complete environmental chemistry AI workflow:

```python
# Complete integrated environmental workflow
print("üîÑ EXECUTING INTEGRATED ENVIRONMENTAL CHEMISTRY AI WORKFLOW")
print("=" * 65)

# Step 1: Multi-media Environmental Monitoring
print("\n1Ô∏è‚É£ Multi-media Environmental Monitoring...")
monitoring_systems = {}
for media in ["air_quality", "water_quality", "soil_contamination"]:
    monitor = EnvironmentalMonitoringSystem(media)
    data = monitor.generate_sample_data(1000)
    metrics = monitor.train_pollution_predictor(data)
    monitoring_systems[media] = {"monitor": monitor, "performance": metrics["r2"]}
    print(f"   ‚úÖ {media.replace('_', ' ').title()}: R¬≤ = {metrics['r2']:.3f}")

# Step 2: Green Chemistry Optimization
print("\n2Ô∏è‚É£ Green Chemistry Process Optimization...")
green_optimizer = GreenChemistryOptimizer()
reaction_data = green_optimizer.generate_reaction_data(600)
optimization_results = green_optimizer.optimize_reaction_conditions(reaction_data)
max_green_score = optimization_results["max_green_score"]
print(f"   ‚úÖ Optimized {len(reaction_data)} reactions, max green score: {max_green_score:.1f}")

# Step 3: Atmospheric Chemistry Analysis
print("\n3Ô∏è‚É£ Atmospheric Chemistry and Air Quality Forecasting...")
atm_analyzer = AtmosphericChemistryAnalyzer()
atm_data = atm_analyzer.generate_atmospheric_data(1500)
trend_analysis = atm_analyzer.analyze_atmospheric_trends(atm_data)
forecasts = atm_analyzer.forecast_air_quality(atm_data)
avg_forecast_accuracy = sum([f["accuracy_r2"] for f in forecasts.values()]) / len(forecasts)
print(f"   ‚úÖ Analyzed {len(atm_data)} atmospheric measurements, forecast accuracy: {avg_forecast_accuracy:.3f}")

# Step 4: Integrated Environmental Assessment
print("\n4Ô∏è‚É£ Generating Integrated Environmental Assessment...")

# Environmental risk assessment
environmental_risk = {
    "air_quality_risk": "low" if forecasts["NO2"]["forecast_value"] < 30 else "moderate",
    "green_chemistry_adoption": "high" if max_green_score > 60 else "moderate",
    "atmospheric_stability": "stable" if abs(trend_analysis["CO2"]["trend_per_hour"]) < 0.001 else "changing"
}

# Sustainability recommendations
sustainability_score = (
    (max_green_score / 100) * 0.4 +  # Green chemistry contributes 40%
    (avg_forecast_accuracy) * 0.3 +   # Forecasting accuracy contributes 30%
    (sum([m["performance"] for m in monitoring_systems.values()]) / len(monitoring_systems)) * 0.3  # Monitoring contributes 30%
)

print("\nüåç INTEGRATED ENVIRONMENTAL ASSESSMENT:")
print(f"   ‚Ä¢ Overall Sustainability Score: {sustainability_score:.3f}")
print(f"   ‚Ä¢ Air Quality Risk Level: {environmental_risk['air_quality_risk'].title()}")
print(f"   ‚Ä¢ Green Chemistry Adoption: {environmental_risk['green_chemistry_adoption'].title()}")
print(f"   ‚Ä¢ Atmospheric Stability: {environmental_risk['atmospheric_stability'].title()}")

# Best performing systems
best_monitor = max(monitoring_systems.items(), key=lambda x: x[1]["performance"])
print(f"   ‚Ä¢ Best Monitoring System: {best_monitor[0].replace('_', ' ').title()} (R¬≤ = {best_monitor[1]['performance']:.3f})")

# Key environmental insights
print("\nüí° KEY ENVIRONMENTAL INSIGHTS:")
print(f"   ‚Ä¢ Pollution prediction models achieve average R¬≤ of {sum([m['performance'] for m in monitoring_systems.values()]) / len(monitoring_systems):.3f}")
print(f"   ‚Ä¢ Green chemistry optimization can achieve sustainability scores up to {max_green_score:.1f}")
print(f"   ‚Ä¢ Atmospheric forecasting provides {avg_forecast_accuracy:.1%} average accuracy")
print(f"   ‚Ä¢ {len([g for g, t in trend_analysis.items() if abs(t['trend_per_hour']) > 0.001])} trace gases show significant trends")

# Recommendations
print("\nüìã RECOMMENDATIONS:")
if sustainability_score > 0.7:
    print("   üåü Excellent environmental management - maintain current practices")
elif sustainability_score > 0.5:
    print("   ‚úÖ Good environmental management - focus on green chemistry optimization")
else:
    print("   ‚ö†Ô∏è Environmental management needs improvement - prioritize pollution monitoring")

if max_green_score < 60:
    print("   üß™ Implement green chemistry protocols to improve sustainability")

if avg_forecast_accuracy < 0.8:
    print("   üìä Enhance atmospheric monitoring network for better forecasting")

print("\n‚úÖ INTEGRATED ENVIRONMENTAL WORKFLOW COMPLETE")
```
